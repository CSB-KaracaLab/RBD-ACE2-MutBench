```python
#import datasets
import pandas as pd
import pathlib
rootdir = pathlib.Path('.').resolve(strict=True)
datadir = rootdir.parents [1]  / 'Output_files'

HADDOCK = pd.read_csv(datadir / 'HADDOCK_Prepared_dataset.csv' ,  delimiter=',')
FoldX = pd.read_csv(datadir / 'FoldX_Prepared_dataset.csv' ,  delimiter=',')
FoldXwater = pd.read_csv(datadir / 'FoldXwater_Prepared_dataset.csv' ,  delimiter=',')
EvoEF1 = pd.read_csv(datadir / 'EvoEF1_Prepared_dataset.csv' ,  delimiter=',')
MutaBind2 = pd.read_csv(datadir / 'MutaBind2_Prepared_dataset.csv' ,  delimiter=',')
SSIPe = pd.read_csv(datadir / 'SSIPe_Prepared_dataset.csv' ,  delimiter=',')
```

```python
# Performance calculation

HADDOCK_performance = HADDOCK['succ_rate'].sum() / len(HADDOCK) * 100

FoldX_performance = FoldX['succ_rate'].sum() / len(FoldX) * 100
EvoEF1_performance = EvoEF1['succ_rate'].sum() / len(EvoEF1) * 100
MutaBind2_performance = MutaBind2['succ_rate'].sum() / len(MutaBind2) * 100
SSIPe_performance = SSIPe['succ_rate'].sum() / len(SSIPe) * 100
FoldXwater_performance = FoldXwater['succ_rate'].sum() / len(FoldXwater) * 100

# Print performance in a file

performances = open("Prediction_performances.txt","a") 
print('HADDOCK prediction performance is : %',round(HADDOCK_performance, 2), file=performances)
print('FoldX prediction performance is : %',round(FoldX_performance, 2), file=performances)
print('EvoEF1 prediction performance is : %',round(EvoEF1_performance, 2), file=performances)
print('MutaBind2 prediction performance is : %',round(MutaBind2_performance, 2), file=performances)
print('SSIPe prediction performance is : %',round(SSIPe_performance, 2), file=performances)
print('FoldXwater prediction performance is : %',round(FoldXwater_performance, 2), file=performances)

performances.close()

#Enriched-Depleted

HADDOCK_Enriched = HADDOCK[HADDOCK['mutation_type'].str.contains("E")]
HADDOCK_Depleted = HADDOCK[HADDOCK['mutation_type'].str.contains("MD") | HADDOCK['mutation_type'].str.contains("RD")]
FoldX_Enriched = FoldX[FoldX['mutation_type'].str.contains("E")]
FoldX_Depleted = FoldX[FoldX['mutation_type'].str.contains("MD") | FoldX['mutation_type'].str.contains("RD")]
EvoEF1_Enriched = EvoEF1[EvoEF1['mutation_type'].str.contains("E")]
EvoEF1_Depleted = EvoEF1[EvoEF1['mutation_type'].str.contains("MD") | EvoEF1['mutation_type'].str.contains("RD")]
MutaBind2_Enriched = MutaBind2[MutaBind2['mutation_type'].str.contains("E")]
MutaBind2_Depleted = MutaBind2[MutaBind2['mutation_type'].str.contains("MD") | MutaBind2['mutation_type'].str.contains("RD")]
SSIPe_Enriched = SSIPe[SSIPe['mutation_type'].str.contains("E")]
SSIPe_Depleted = SSIPe[SSIPe['mutation_type'].str.contains("MD") | SSIPe['mutation_type'].str.contains("RD")]
FoldXwater_Enriched = FoldXwater[FoldXwater['mutation_type'].str.contains("E")]
FoldXwater_Depleted = FoldXwater[FoldXwater['mutation_type'].str.contains("MD") | FoldXwater['mutation_type'].str.contains("RD")]

# Performance calculation

HADDOCK_Enriched_performance = HADDOCK_Enriched['succ_rate'].sum() / len(HADDOCK_Enriched) * 100
HADDOCK_Depleted_performance = HADDOCK_Depleted['succ_rate'].sum() / len(HADDOCK_Depleted) * 100
FoldX_Enriched_performance = FoldX_Enriched['succ_rate'].sum() / len(FoldX_Enriched) * 100
FoldX_Depleted_performance = FoldX_Depleted['succ_rate'].sum() / len(FoldX_Depleted) * 100
EvoEF1_Enriched_performance = EvoEF1_Enriched['succ_rate'].sum() / len(EvoEF1_Enriched) * 100
EvoEF1_Depleted_performance = EvoEF1_Depleted['succ_rate'].sum() / len(EvoEF1_Depleted) * 100
MutaBind2_Enriched_performance = MutaBind2_Enriched['succ_rate'].sum() / len(MutaBind2_Enriched) * 100
MutaBind2_Depleted_performance = MutaBind2_Depleted['succ_rate'].sum() / len(MutaBind2_Depleted) * 100
SSIPe_Enriched_performance = SSIPe_Enriched['succ_rate'].sum() / len(SSIPe_Enriched) * 100
SSIPe_Depleted_performance = SSIPe_Depleted['succ_rate'].sum() / len(SSIPe_Depleted) * 100
FoldXwater_Enriched_performance = FoldXwater_Enriched['succ_rate'].sum() / len(FoldXwater_Enriched) * 100
FoldXwater_Depleted_performance = FoldXwater_Depleted['succ_rate'].sum() / len(FoldXwater_Depleted) * 100

# Print performance in a file

performances = open("Prediction_performances.txt","a") 
print('HADDOCK Enriched prediction performance is : %',round(HADDOCK_Enriched_performance, 2), file=performances)
print('FoldX Enriched prediction performance is : %',round(FoldX_Enriched_performance, 2), file=performances)
print('EvoEF1 Enriched prediction performance is : %',round(EvoEF1_Enriched_performance, 2), file=performances)
print('MutaBind2 Enriched prediction performance is : %',round(MutaBind2_Enriched_performance, 2), file=performances)
print('SSIPe Enriched prediction performance is : %',round(SSIPe_Enriched_performance, 2), file=performances)
print('FoldXwater Enriched prediction performance is : %',round(FoldXwater_Enriched_performance, 2), file=performances)

print('HADDOCK Depleted prediction performance is : %',round(HADDOCK_Depleted_performance, 2), file=performances)
print('FoldX Depleted prediction performance is : %',round(FoldX_Depleted_performance, 2), file=performances)
print('EvoEF1 Depleted prediction performance is : %',round(EvoEF1_Depleted_performance, 2), file=performances)
print('MutaBind2 Depleted prediction performance is : %',round(MutaBind2_Depleted_performance, 2), file=performances)
print('SSIPe Depleted prediction performance is : %',round(SSIPe_Depleted_performance, 2), file=performances)
print('FoldXwater Depleted prediction performance is : %',round(FoldXwater_Depleted_performance, 2), file=performances)

performances.close()
```
